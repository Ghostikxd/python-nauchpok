```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import category_encoders as ce

df = pd.read_csv('2023-11-07_Opros_pervokursnikov_3.csv')  

features = df.columns[3:]

encoder = ce.CountEncoder()
df_encoded = encoder.fit_transform(df[features])

tsne = TSNE(n_components=2, random_state=42)
tsne_representation = tsne.fit_transform(df_encoded)

df_tsne = pd.DataFrame(tsne_representation, columns=['Dimension 1', 'Dimension 2'])

plt.figure(figsize=(10, 8))
plt.scatter(df_tsne['Dimension 1'], df_tsne['Dimension 2'], cmap='viridis')
plt.title('t-SNE Visualization')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.colorbar()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('2023-11-07_Opros_pervokursnikov_3.csv')  

features = df.columns[3:]

tsne = TSNE(n_components=2, random_state=42)

df_tsne = pd.DataFrame(columns=['Dimension 1', 'Dimension 2'])

for column in features:
    encoder = LabelEncoder()
    df_tsne[column] = encoder.fit_transform(df[column])

tsne_representation = tsne.fit_transform(df_tsne[features])

df_tsne = pd.DataFrame(tsne_representation, columns=['Dimension 1', 'Dimension 2'])

plt.figure(figsize=(10, 8))
plt.scatter(df_tsne['Dimension 1'], df_tsne['Dimension 2'], cmap='viridis')
plt.title('t-SNE Visualization')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.colorbar()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

df = pd.read_csv('2023-11-07_Opros_pervokursnikov_3.csv')  

# Автоматическое кодирование категориальных признаков
df_encoded = pd.get_dummies(df, drop_first=True)

tsne = TSNE(n_components=2, random_state=42)
tsne_representation = tsne.fit_transform(df_encoded)

plt.figure(figsize=(10, 8))
plt.scatter(tsne_representation[:, 0], tsne_representation[:, 1], cmap='viridis')
plt.title('t-SNE Visualization')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.colorbar()
plt.show()
```
